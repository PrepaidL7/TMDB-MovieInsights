__The next functions make an HTTP request to the TMDb API to retrieve movie results for a specific page within a date range and utilize parallelism with `ThreadPoolExecutor` to fetch multiple pages concurrently, while handling connection errors or request failures.__

```python
import time
from concurrent.futures import ThreadPoolExecutor

start = time.time()

def fetch_page(page):
    params = {'api_key': '2718ed700b242610b03e00cf6f87a949',
              'primary_release_date.gte': '1874-12-09',
              'primary_release_date.lte': '2017-12-27',
              'page': page}
    try:
        response = requests.get(url, params=params, timeout=40) 
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error en la página {page}: {e}")
        return None


def fetch_pages_with_wait(pages, batch_size=25, wait_time=9):
    completed_count = 0
    results = []
    
    with ThreadPoolExecutor(max_workers=20) as executor:
        for result in executor.map(fetch_page, pages):
            if result is not None:
                results.append(result)
            completed_count += 1

            
            if completed_count % batch_size == 0:
                print(f"{completed_count} solicitudes completadas, esperando {wait_time} segundos...")
                time.sleep(wait_time)

    return results

pages = range(1, 11)

results = fetch_pages_with_wait(pages)

print(f"El loop tomó {time.time() - start:.2f} segundos.")
```
